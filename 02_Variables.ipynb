# Variables in Python

### What is variables?
- A variable is used to store data values.

### Example

```python
name = "Kavya kashyap"
age = 20
print(name)
print(age)
```

### Variable Naming Rules
- Must start with a letter or ( _ )
- Can not start with a number
- Variable name can not be the same
- Python is a case-sensetive language so (name) and (NAME) would be considered different variables.

### Valid Variable Names

- name = "Kavya kashyap"
- _age = 20
- student_name = "Kavya kashyap"

### Invalid Variable Names

- 2name = "Kavya kashyap"
- student-name = "Kavya kashyap"
