# Data Types in Python

 - Data types are used to define the type of data that a variable can hold. They help python to understand how data should be stored and processed.

# Common Data Types

### None (none)

```python
x = None
print(x)
print(type(x))
```

 ### Integer (int)

```python
age = 20
print(age)
print(type(age))
```

### Float (float)

```python
height = 5.2
print(height)
print(type(height))
```

### String (str)

```python
name = "Kavya kashyap"
print(name)
print(type(name))
```

### Boolean (bool)

```python
is_student = True
print(is_student)
print(type(is_student))
```

### List (list)

```python
List = [1,2,3,4,5]
print(List)
print(type(List))
```

### Tuple (tuple)

```python
Names = ("Kavya","Rohan","Shagun","Raaj")
print(Names)
print(type(Names))
```

### Dictionary (dict)

```python
student = {"name":"Kavya", "age":20}
print(student)
print(type(student))
```

### Set (set)

```python
colors = {"Green","Blue","Red","Yellow"}
print(colors)
print(type(colors))
```
