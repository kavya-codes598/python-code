### Basic Operations
- Concatenation

```python
string1 = "Kavya"
string2 = "Kashyap"
final_string = string1 +" "+ string2
print(final_string)
```

- Lenth of string

```python
string1 = "Kavya"
print(len(string1))

string2 = "Kashyap"
print(len(string2))

final_string = string1 +" "+ string2
print(final_string)
print(len(final_string))
```

### Indexing 

```python
str1 = "Python code"
b = str1[0]
c = str1[3]
d = str1[5]
print(b,c,d)
# str1[0] = x is not allowed
```

### Slicing : [Accessing the part of sting ]

```python
str1 = "Python code"
print(str1[1:5])
print(str1[:8]) #[0:8]
print(str1[0:]) #[0:len(str1)]
```

### Negative Indexing

```python
str2 = "Apple"
print(str2[-3:-1])
print(str2[-5:-1])
print(str2[-1:-4])
```

## Conditional Statments ( if, elif , else)

- If

```python
age = 23
if age>=18:
    print("Can vote")
    print("Can drive")
```

- elif

```python
age = 16
if age>=18:
    print("Can vote")
    print("Can drive")
elif age<=18:
    print("Can not vote")
    print("Can not drive")
```

- else

```python
age = 15
if age>=18:
    print("Can vote")
    print("Can drive")
else:
    print("You are not eligible")
```

- Example of ( if, elif ,  else )

```python
light = "black"

if light=="red":
    print("Stop")
elif light=="yellow":
    print("Ready")
elif light=="green":
    print("Go")
else: 
    print("Light is broken")
```

- Geade based on marks

```python
Student_marks = int(input("Enter the marks of student: "))
if Student_marks>=90:
    print("Grade: 'A'")
elif 90>Student_marks>=80:
    print("Grade: 'B'")
elif 80>Student_marks>=70:
    print("Grade: 'C'")
elif 70>Student_marks>35:
    print("Grade: 'D'")
else:
    print("Fail")
```
