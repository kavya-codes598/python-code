# Logical Operator

```python
print(user_name==user_name_by_user and password==password_by_user)
```

#### 2. Logical ( or Operator )

```python
print(user_name==user_name_by_user or password==password_by_user)
```

# Identity Operator [ is , is not ]

#### 1. is  Operator

```python
a = 10
b = 10
print(a is b)
```

```python
a = [1,2,3]
b = [1,2,3]
print(a is b)
```

#### 2. is not Operator

```python
a = 10
b = 13
print(a is not b)
print(id(a))
print(id(b))
```

```python
a = [1,2,3]
b = [1,2,3]
print(a is b)
print(id(a))
print(id(b))
```
