## List in Python []
- A built-in data type that stores set of values

```python
mark1 = 94.1 # A single variable
mark2 = 84.2 # A single variable
mark3 = 76.5 # A single variable
mark4 = 56.2 # A single variable
mark5 = 45.1 # A single variable

marks = [94.1, 84.2, 76.5, 56.2, 45.1] # list
print(marks)
print(type(marks))
```

- Access through index value 

```python
marks[0]
```

```python
marks[2]
```

```python
marks[4]
```

- It can store elements of  different types ( int, float, string, etc )

```python
student = [ "Kavya", 94.1, 20, "Noida"]
print(student)
print(type(student))
print(len(student))
```

- It is mutable

```python
student = [ "Kavya", 94.1, 20, "Noida"]
student[0] = "kashyap"
print(student)
```

## List Slicing 
- It is similer to string
- list_name[staring_idx : ending_idx]  #ending idx not included


```python
marks =[74, 84, 67, 85, 90]
print(marks[0:4])
print(marks[:3])
print(marks[1:])
```

- negative indexing

```python
print(marks[-3:-1])
print(marks[-4:-2])
```

## List Methods

```python
list = [5, 3, 8, 4, 1, 6]
```

- list.append() --> adds one element at the end

```python
list.append(10)
list
```

- list.sort() --> sort in ascending order

```python
list.sort()
print(list)
```

- list.sort(reverse = True)-->sort in descending order

```python
list.sort(reverse = True)
print(list)
```

```python
list1 = ['f', 'd', 'a', 'c', 'b', 'e']
list1.sort()
list1
```

- list.reverse() --> reverse list

```python
list.reverse()
print(list)
```

- list.insert(idx,el)

```python
list.insert(2,2)
print(list)
```

- list.remove() --> remove first occurrence element

```python
list.remove(2)
print(list)
```

- list.pop(idx)--> remove element at idx

```python
list.pop(4)
print(list)
```

## Tuples in Python ()
- A built-in data type that lets us create immutable sequence of values

```python
tup = (1, 2, 3, 4, 5, 6, 7, 8)
print(tup)
print(len(tup))
print(type(tup))
```

- Access through indexing

```python
tup[0]
```

```python
tup[2]
```

```python
 # tup[2] = 5 is not allowed (immutable)
```

## Slicing in Tuple

```python
tup[2:5]
```

```python
tup[:7]
```

```python
tup[2:]
```
