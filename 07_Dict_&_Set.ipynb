## Dictionary in Python { key:value }
- Dictionary are used to store data values in key:value pairs.
- They are unordered, mutable(changeable) & don't allow duplicate keys.

```python
dict = {
    "key" : "value",
    "name" : "kavya",
    "age" : 20,
    "subjects" : ["python", "Django", "java"],
    "marks" : (89, 56, 54, ),
    12.5:89.5
}
print(dict)
print(type(dict))
```

```python
dict["name"] = "Kavya Kashyap"
print(dict["name"])
```

```python
print(dict["subjects"])
```

```python
null_dict = {}
print(null_dict)
```

```python
null_dict["learning"] = "Python"
null_dict
```

## Nested Dictionaries

```python
student = {
    "name":"Kavya Kashyap",
    "subjects":{
        "math":89,
        "Eng":85,
        "Phy":90,
    },
}
print(student)
print(student["subjects"] )
print(student["subjects"] ["Eng"])
```

## Set in Python 
- set is collection of the unordered items.
- Each item in the set must be unique & immutable.

```python
set0 = {1, 2, 3, 4, 5,}
print(set0)
print(len(set0))
print(type(set0))
```

- set ignore dublicates

```python
set1 = {1, 1, 2, 2, "hello", "hello", 8, }
print(set1)
```

```python
print(len(set1))
print(type(set1))
```

```python
collection = set()
print(type(collection))
print(collection)
```

## Set Methodes

- set.add()

```python
my_set = set()
my_set.add(1)
my_set.add(2)
my_set.add(2)
my_set.add(1)
my_set.add(("name", "Pyhton", 23, 79.2,))
print(my_set)
```

- set.pop()

```python
my_set.pop()
print(my_set.pop())
print(my_set)
```

```python
set2 ={1, 2, 3, 4, 34,"kavya", "book", "kashyap"}
print(set2.pop())
print(set2.pop())
print(set2.pop())
print(set2.pop())
```

- set.union()

```python
num = {1, 2, 3, 4, }
num2 = {3, 4, 5, 6,}
new_num = num.union(num2)
print(new_num)
```

- set.intersection()

```python
num = {1, 2, 3, 4, }
num2 = {3, 4, 5, 6,}
print( num.intersection(num2))
```
